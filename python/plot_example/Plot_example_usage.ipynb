# 热力图

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt


def h(x, y):  
    return (1-x/2+x**5+y**3) * np.exp(-x**2-y**2) 

 

m,n=200,250
x=np.linspace(-3,3,m)
y=np.linspace(-3,3,n)

 

X,Y=np.meshgrid(x,y)

 

plt.contourf(X,Y,h(X,Y))
plt.contour(X,Y,h(X,Y),10)

 

plt.show()        
```
