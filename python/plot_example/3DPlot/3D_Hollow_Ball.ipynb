## 3D scatter Plot

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)

c = x + y

ax.scatter(x, y, z, c=c)
```

## 3D Line Plot

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot(x, y, z, '-b')
```

## Surface Plot

```python
x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
y = x.copy().T
z = np.cos(x ** 2 + y ** 2)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, z, cmap=plt.cm.jet, rstride=1, cstride=1, linewidth=0)
```

## Wireframe plot

```python
u = np.linspace(0, np.pi, 20)
v = np.linspace(0, 2 * np.pi, 20)

x = np.outer(np.sin(u), np.sin(v))
y = np.outer(np.sin(u), np.cos(v))
z = np.outer(np.cos(u), np.ones_like(v))

single_col_width=2.5
single_row_height=2.5
figure_dpi=600
fig = plt.figure(figsize=(single_col_width, single_row_height),dpi=figure_dpi)
ax = plt.axes(projection='3d')

# ax.plot_wireframe(x, y, z,color='#000000',linestyle=':')
ax.plot_wireframe(x, y, z,color='#000000')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
plt.axis('off')
# ax.set_facecolor()
ax.grid(False)
save_dir=r"E:\Users\scfan\data\image_project\fill_NAN\without_nan_data\measure_of_sequence_matching_2021_09_28\fig"
save_filename="3d_hollow_ball.png"
save_path=os.path.join(save_dir,save_filename)
plt.savefig(save_path, transparent=True)
```
